# **Seperating Dependent & Independent Variables**

---



```python
X = df.drop(["id","outcome"], axis=1)
y = df["outcome"]
```

```python
X.head()
```

```python
y.head()
```

# **Encoding**

---



```python
from sklearn.preprocessing import LabelEncoder
```

```python
cols  = [i for i in X.columns]
types = [i for i in X.dtypes]
```

```python
cat_cols=list()
enc=dict()

for i in range(len(types)):

  if(types[i]==object):

    cat_cols.append(cols[i])
    enc[cols[i]] = LabelEncoder()
    X[cols[i]] = enc[cols[i]].fit_transform(X[cols[i]])
```

```python
X.head()
```

```python
o_le = LabelEncoder()
```

```python
y = o_le.fit_transform(y)
```

```python
y_map = dict(zip(o_le.classes_, o_le.transform(o_le.classes_)))
```

```python
y_map
```

```python
o_map=dict()
for i in y_map:

  o_map[y_map[i]] = i
```

```python
o_map
```

# **Splitting into Training & Testing Set**

---



```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                    random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# **Model Building - Decision Tree Classifier**

---



```python
from sklearn.tree import DecisionTreeClassifier
dtc_model = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
dtc_params = {'max_depth': list(range(10, 15)),
               'max_features': list(range(2,14))}
```

```python
dtc_cv = RandomizedSearchCV(estimator=dtc_model,
                            param_distributions=dtc_params,
                            cv=10)
```

```python
dtc_cv.fit(X_train, y_train)
```

```python
dtc_cv.best_params_
```

```python
dtc_model = DecisionTreeClassifier(max_features=8, max_depth=10)
```

```python
dtc_model.fit(X_train, y_train)
```

```python
dtc_model.score(X_test, y_test)
```

```python
y_preds1 = dtc_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_preds1))
```

# **Model Building - XGBClassifier**

---



```python
from xgboost import XGBClassifier
xgbc_model = XGBClassifier()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
xgbc_params = {'gamma' :[i/10.0 for i in range(0,5)],
               'max_depth' : [2,4,6,8,10,12,13,15],
               'min_child_weight':[1,2,5,10],
               'max_delta_step':[0,1,2,5,10],
               'subsample' :[i/10.0 for i in range(5,10)],
               'colsample_bytree' :[i/10.0 for i in range(5,10)],
               'colsample_bylevel' :[i/10.0 for i in range(5,10)],
               'n_estimators' : [100,150,200,300]
               }
```

```python
xgbc_cv = RandomizedSearchCV(estimator=xgbc_model,
                            param_distributions=xgbc_params,
                            cv=10)
```

```python
xgbc_cv.best_params_
```

```python
xgbc_model = XGBClassifier( subsample = 0.8,
                           n_estimators = 100,
                            min_child_weight = 5,
                            max_depth = 4,
                            max_delta_step = 2,
                            gamma = 0.4,
                            colsample_bytree = 0.8,
                            colsample_bylevel = 0.6
                            )
```

```python
xgbc_model.fit(X_train, y_train)
```

```python
xgbc_model.score(X_test, y_test)
```
